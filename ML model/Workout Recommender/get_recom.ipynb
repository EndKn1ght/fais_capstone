# Dependencies

```python
import pandas as pd
import pickle
from sklearn import metrics
```

# Load encodings

```python
def load_encodings():
  file_path = 'encodings.pickle'
  with open(file_path, 'rb') as pkl:
    encodings = pd.DataFrame(pickle.load(pkl))
  return encodings
```

```python
encodings = load_encodings()
```

# Retrieving top 3 recommendation

```python
def get_recommendations(workout_index):
  sim_score = pd.DataFrame()
  sim_score['similarity_score'] = encodings['encodings'].apply(lambda x: metrics.pairwise.cosine_similarity(x, encodings.iat[workout_index, 0])[0][0])
  sim_score = sim_score.sort_values(by=['similarity_score'], ascending=False)
  sim_score = sim_score.drop(workout_index, axis=0)
  top3 = sim_score.iloc[0:3]
  return top3.index
```

```python
get_recommendations(int(input()))
```
